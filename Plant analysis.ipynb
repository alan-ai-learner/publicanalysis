# Awesome Public Analysis
datasource: https://www.plants.usda.gov/dl_all.html

```python
import pandas as pd
```

```python
df = pd.read_csv('plants.txt', sep=",")
```

```python
df.head(n=25)
```

```python
count_df = df.groupby(['Family']).size().reset_index(name='counts')
```

```python
count_df.sort_values(by=['counts'], ascending=False)
```

```python
# Import wiki, summarize wiki page
```

```python
import wikipedia
page = wikipedia.page('Asteraceae')
```

```python
print len(page.images)
```

```python
from IPython.display import Image, SVG, display
def show_svg(url):
    display(Image(url=url,width=300, height=300))
```

```python
page.images[3]
```

```python
wikipedia.summary('Asteraceae', sentences=5)
```

```python
import matplotlib.pyplot as plt
plt_data = count_df[count_df['counts'] > 1000]
```
